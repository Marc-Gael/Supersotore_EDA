### Clustering
Identifier les clients qui font uniquement des achats en discount

### Préparation des données

```python

df_cleaned["Promo_Purchase"] = df_cleaned["Discount"].apply(lambda x: 1 if x > 0 else 0)
df_cleaned.head()
```

```python
df_cleaned.head()
```

```python
df_cleaned.rename(columns={"promo_purchase": "Promo_Purchase"}, inplace=True)

df_cleaned.head()
```

```python
# Agrégation des données par code postal
postal_code_data = df_cleaned.groupby('Postal Code').agg(
    Total_Orders=('Sales', 'count'),
    Promo_Orders=('Promo_Purchase', 'sum'),
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()

postal_code_data['Promo_Order_Ratio'] = postal_code_data['Promo_Orders'] / postal_code_data['Total_Orders']
postal_code_data
```

### Application du clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

 # Sélection des variables pour le clustering
X = postal_code_data[['Promo_Order_Ratio', 'Total_Sales', 'Total_Profit']]

 # Normalisation des données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

 # Application de K-means
kmeans = KMeans(n_clusters=3, random_state=42)
postal_code_data["Cluster"] = kmeans.fit_predict(X_scaled)
```

### Analyse des clusters

```python
# Résumé des clusters
cluster_summary = postal_code_data.groupby('Cluster').agg(
    Avg_Promo_Ratio=('Promo_Order_Ratio', 'mean'),
    Avg_Total_Sales=('Total_Sales', 'mean'),
    Avg_Total_Profit=('Total_Profit', 'mean'),
    Count=('Postal Code', 'count')
).reset_index()

# Filtrage des codes postaux où les achats sont faits uniquement lors des promos
promo_only_postal_codes = postal_code_data[postal_code_data['Promo_Order_Ratio'] == 1]
promo_only_postal_codes
```
